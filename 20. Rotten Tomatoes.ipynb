# Rotten Tomatoes
http://www.rottentomatoes.com/top/bestofrt/?year=2019

```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
import time
import numpy as np

# HTML 가져오기
url = "http://www.rottentomatoes.com/top/bestofrt/?year=2019"
res = req.urlopen(url)

# HTML 분석하기
soup = BeautifulSoup(res, "html.parser")
```

### Rank 테스트


```python
rank_Datas = soup.select("table.table td.bold")
rank_Lists = []

for rank in rank_Datas:
    rank_Lists.append(rank.string.replace(rank.string, rank.string.split('.')))
    
rank_Lists
```

### Rating 테스트

```python
rating_Datas = soup.select("span.tMeterScore")
rating_Lists = []

for rating in rating_Datas:
    rating_Lists.append(rating.string.strip())
    
rating_Lists
```

### Title 테스트

```python
li_lists = soup.select("#top_movies_main > div > table > tr > td > a")
count = 0
titleLists = [] # 영화제목 List
countLists = [] # Count List

for li in li_lists:
    count += 1
    countLists.append(count)
    titleLists.append(li.string.strip())
    
titleLists
```

### Review 테스트

```python
review_Datas = soup.select("td.right")
review_Lists = []

for review in review_Datas:
    review_Lists.append(review.string.strip())
    
review_Lists
```

```python

# Title 선택후 Genres 가져오기

li_titles = soup.select("#top_movies_main > div > table > tr > td > a")
urlBase = 'https://www.rottentomatoes.com'
titleLists = []
genreLists = []
urlLists = []
count1 = 0

for li in li_titles:
    titleLists.append(li.string.strip())
    genreStr = ""
    workUrl = urlBase + li['href']
    pageSelect = req.urlopen(workUrl)
    soupSelect = BeautifulSoup(pageSelect, 'html.parser')
    try:
        genreStr = soupSelect.select("li > div.meta-value.genre")[0].string.replace(",", "").strip().replace(" ","").split()
    except IndexError as e:
        genreLists.append(np.NaN)
        urlLists.append(workUrl)
        count1 += 1
    else:
        genreLists.append(genreStr)
        urlLists.append(workUrl)
        count1 += 1
        
    print(">>>>>", str(count1))
    print("Title :", li.string.strip())
    print("Genre :", genreStr)
    print("URL :", workUrl)
    print("----------------------------------------------------")
    time.sleep(1)
```

```python
## 쌤이 하신 거 !!
# Rank
li_ranks = soup.select("table.table td.bold")
# Rating
li_ratings = soup.select("span.tMeterScore")
#print(li_ratings)
# Titles
li_titles = soup.select("#top_movies_main > div > table > tr > td > a")
# Reviews
li_reviews = soup.select("td.right")

rankLists = []
ratingLists = []
titleLists = []
reviewLists = []

# Rank 가져오기
for li in li_ranks:
    rankT = li.string.strip()
    rankT = rankT.replace(".","")
    rankLists.append(rankT)
    
# Rating 가져오기
for li in li_ratings:
    ratingLists.append(li.string.strip())
    
print(ratingLists)

# Title 가져오기
for li in li_titles:
    titleLists.append(li.string.strip())
    
# Review 가져오기
for li in li_reviews:
    reviewLists.append(li.string.strip())
    
```

```python
# List 확인
print(rankLists)
print(ratingLists)
print(titleLists)
print(reviewLists)
print(genreLists)
print(urlLists)
```

```python
# csv 만들기
df = pd.DataFrame(data=rankLists, columns=['Rank'])
df['Rating'] = pd.Series(ratingLists)
df['Title'] = pd.Series(titleLists)
df['Review'] = pd.Series(reviewLists)
df['Genre'] = pd.Series(genreLists)
df['URL'] = pd.Series(urlLists)
df
```

```python
# csv로 저장
df.to_csv("../Data/rottenTomatoesList_2019.csv", encoding='utf-8', index=False)
```
