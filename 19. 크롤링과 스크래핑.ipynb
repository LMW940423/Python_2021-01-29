# 데이터 다운로드 하기 - 2
: 인터넷에서 지정된 파일을 메모리에 올리고 저장하기

```python
import urllib.request

# URL과 저장경로 지정하기.
url = "http://uta.pw/shodou/img/28/214.png"
savename = "../Data/test02.png"

# 다운로드 하기
mem = urllib.request.urlopen(url).read()

# 파일로 저장하기
with open(savename, "wb") as f: # 텍스트냐 binary냐에 따라 명령어가 다르다.  
    f.write(mem)
    
print("저장 되었습니다!")
```

# 매개변수를 추가해 요청을 전송하는 방법
기상청 RSS : http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp    
URL 뒤에 매개변수를 입력하여 지역별로 일기 정보를 가져올 수 있음.   
전국 : 108, 서울/경기도 : 109, 강원도 : 105, 제주도 : 184   

ex) 서울/경기도 : http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stdId=109

```python
import urllib.request
import urllib.parse

API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

# 매개변수를 URL Encoding (?stdid=019 => dictionary)
values = {'stnId':'109'}
params = urllib.parse.urlencode(values) # 인코딩한 내용을 parse한다

# 요청 전용 URL 생성
url = API + "?" + params
print("url = ", url)

data = urllib.request.urlopen(url).read()
#print(data) # 글자를 확인할 수 없음
text = data.decode("utf-8")
print(text)
```

# BeautifulSoup로 Scraping하기
- BeautifulSoup : 간단하게 HTML과 XML에서 정보 추출
- BeautifulSoup 설치 : pip install beautifulsoup4

```python
# 기본 사용법
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <h1> Header </h1>
                <p> Line 1을 서술 </p>
            </body>
        <html>
"""

# HTML 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
h1 = soup.html.body.h1
p = soup.html.body.p

print(h1)
print(p)

# text만 출력
print("h1 =", h1.string)
print("p =", p.string)
```

```python
# id로 요소 추출하기

from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <h1 id = "title"> Header </h1>
                <p id = "body"> Line 1을 서술 </p>
            </body>
        <html>
"""

# HTML 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
title = soup.find(id='title')
body = soup.find(id='body')
print(title)
print(body)

# text만 출력
print(title.string)
print(body.text)
```

```python
# 여러개의 요소 추출하기
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <ul>
                    <li><a href="http://www.naver.com">naver</a></li>
                    <li><a href="http://www.daum.net">daum</a></li>
                </ul>
            </body>
        <html>
"""
# HTML 분석
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기
links = soup.find_all('a') # find_all은 모두 추출 / find는 처음 하나만 추출
print(links)

for link in links:
    href = link.attrs['href'] # a태그안의 href값
    text = link.string # a태그의 name값
    print(text, "==>", href)
```

```python
# 기상청 자료 활용하기
from bs4 import BeautifulSoup
import urllib.request as req

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"

# data 가져오기
res = req.urlopen(url)

# BeautifulSoup으로 분석
soup = BeautifulSoup(res, 'html.parser')
#print(soup) # 전체 데이터 확인

# 원하는 데이터 추출하기
title = soup.find('title').string # 제일 첫 태그값만 가져온다
wf = soup.find('wf').string

print(title)
print(wf)

# <br />제거하고 정렬하기 (다시풀기!!!)
#list_wfs = list(wf)

except_char = ['<', 'b', 'r', '/', '>']
result = ''

for lwf in wf:
    if lwf not in except_char:
        result += lwf
    if lwf == '○':
        wf.replace('○', '\n ○')
        result += lwf
        
print("=" * 50)        
print(result)
```

# CSS 선택자 사용하기
- soup.select_one() : CSS 선택자로 요소 하나를 추출
- soup.select() : CSS선택자로 요소 리스트를 추출

```python
# 여러개의 요소 추출하기
from bs4 import BeautifulSoup

# HTML Sample
html = """
        <html>
            <body>
                <div id = "meigen">
                    <h1>위키 북스</h1>
                    <ul class="items">
                        <li>유니티 게임 이팩트 입문서</li>
                        <li>스위프트로 시작하는 아이폰 앱 개발 교과서</li>
                        <li>모던 웹사이트 디자인의 정석</li>
                    </ul>
                </div>
            </body>
        <html>
"""

# HTML 분석하기
soup = BeautifulSoup(html, 'html.parser')

# 필요한 부분 css로 추출하기
# 타이틀 부분 추출하기
h1 = soup.select_one("div#meigen > h1").string # id:# / class:. / tag:tag명 / 자손:> / 후손:' '
print("h1 :",h1)

# 목록 부분 추출하기
li_lists = soup.select("ul.items > li")
#li_lists = soup.select("div#meigen  > ul.items > li") # 위에 경우 안먹힐 경우!

for li in li_lists:
    print("li :", li.string)
```

# 네이버 금융에서 환율 정보 추출하기

https://finance.naver.com/marketindex/

```python
# 미국 환율 가져오기
from bs4 import BeautifulSoup
import urllib.request as req

# HTML 가져오기
url = "https://finance.naver.com/marketindex/"
res = req.urlopen(url)

# HTML 분석하기
soup = BeautifulSoup(res, "html.parser")

# 데이터 추출하기
price = soup.select_one("div.head_info > span.value").string
print(price)
```

```python
# 미국, 일본, 유럽연합, 중국의 환율 구하기
from bs4 import BeautifulSoup
import urllib.request as req

# HTML 가져오기
url = "https://finance.naver.com/marketindex/"
res = req.urlopen(url)

# HTML 분석하기
soup = BeautifulSoup(res, "html.parser")

# 데이터 추출하기
USD = soup.select_one("a.usd > div.head_info > span.value").string
EUR = soup.select_one("a.eur > div.head_info > span.value").string
JPY = soup.select_one("a.jpy > div.head_info > span.value").string
print(USD + '원')
print(EUR + '원')
print(JPY + '원')

prices = soup.select("div.head_info > span.value")
print(prices)

print("미국 :", prices[0].string)
print("일본 :", prices[1].string)
print("유럽연합 :", prices[2].string)
```

```python
from bs4 import BeautifulSoup
import urllib.request as req

# HTML 가져오기
url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
res = req.urlopen(url)

# HTML 분석하기
soup = BeautifulSoup(res, "html.parser")

# 데이터 추출하기
a_list = soup.select("#mw-content-text > div.mw-parser-output > ul > li a")

for a in a_list:
    name = a.string
    if name == '증보판':
        continue
    print(name)
```

# Rotten Tomatoes (csv파일로 만들기)
http://www.rottentomatoes.com/top/bestofrt/?year=2019

```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

# HTML 가져오기
url = "http://www.rottentomatoes.com/top/bestofrt/?year=2019"
res = req.urlopen(url)

# HTML 분석하기
soup = BeautifulSoup(res, "html.parser")
```

### 테스트

```python
#top_movies_main > div > table > tbody > tr:nth-child(1) > td:nth-child(3) > a
soup.select("#top_movies_main > div > table > tr > td > a")[0].string
```

## 적용

```python
li_lists = soup.select("#top_movies_main > div > table > tr > td > a")
count = 0
titleLists = [] # 영화제목 List
countLists = [] # Count List

for li in li_lists:
    count += 1
    print(count, li.string.strip())
    countLists.append(count)
    titleLists.append(li.string.strip())
```

```python
# Lists 확인
print(countLists)
print(titleLists)
```

```python
# countList만으로 dataframe 생성
df = pd.DataFrame(data=countLists, columns=['index'])
df
```

```python
# 기존 dataframe에 titleLists 추가
df['Titles'] = pd.Series(titleLists)
df
```

```python
# csv로 저장
df.to_csv("../Data/rottenTomatoesList_2019.csv", encoding='utf-8', index=False)
```
